```python
#String- Task 1

s = input()

count1 = 0
count2 = 0

for i in s:
  if i >= 'a' and i <= 'z':
    count1 = count1 + 1
  elif i >= 'A' and i <= 'Z':
    count2 = count2 + 1


if count2 > count1:
  print(s.upper())

else:
  print(s.lower())
```

```python

# Task-2

t=[]
for i in range(3):
  t.append(input())


for s in t:
  num=False
  ch=False

  for i in s:
    if 65<=ord(i)<=90 or 97<=ord(i)<=122:
      ch=True
    else:
      num=True
    if ch == True and num == True:
      break
  if ch==True and num==False:
    print("WORD")
  elif ch==False and num==True:
    print("NUMBER")
  else:
    print("MIXED")
```
