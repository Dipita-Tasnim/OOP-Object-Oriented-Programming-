```python
#Task 1 

class Exam:
  def __init__(self, mark):
    self.mark = mark

  def __add__(self, other):
    self.other = other #eta na likhleo hoy karon self.other amar lagtese na.

    total = self.mark + other.mark
    return Exam(total)

Q1 = Exam(int(input("Quiz 1 (out of 10): ")))
Q2 = Exam(int(input("Quiz 2 (out of 10): ")))
Lab = Exam(int(input("Lab (out of 30): ")))
Mid = Exam(int(input("Mid (out of 20): ")))
Final = Exam(int(input("Final (out of 30): ")))
total = Q1 + Q2 + Lab + Mid + Final
print("Total marks: {}".format(total.mark))

```

```python
# Task 6 

# return e comma diye duita jinish return koray auto kivabe duita method e gelo
# karon- jehetu operator overload wala method tai return e er por Class er nam 
# thakbe. Aar class er nam diye return kora mane sheta constractor e giye receive
# hobe. Orthat ekhon ei duita val self.__base aar self.__height e ache. ebar
# jokhon getBase method call hobe tokhon self.__base ta jeye print hobe.Abar
# jokhon getHeight method call hobe tokhon self.__height ta jeye print hobe.  


class Triangle:
  def __init__(self, base, height):
    self.__base = base
    self.__height = height
    #self.__area = 0  #eta dorkar nai


  def getBase(self):
    return self.__base 


  def getHeight(self):
    return self.__height


  def area(self):
    self.__area = 0.5 * self.__base * self.__height
    return self.__area


  def __sub__(self, other): # sub method ektar moddhei duitar kaj hobe. parameter o ekta.
    self.__differenceB = (self.__base - other.__base)
    self.__differenceH = (self.__height - other.__height)
    return Triangle(self.__differenceB, self.__differenceH)


  # def setBase(self, subBase):  #ei duita method na likhleo kaj korbe.karon
  #   self.__base = self.__differenceB #submethod e already return kore 
  #                                            #constructor e pathano hoise.
  #                                             #aar setmethod driver code e call o
  #                                                #deya hoy nai.
  # def setHeight(self, subHeight):    
  #   self.__height = self.__differenceH



t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())
 
t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
 
t3 = t1 - t2 
print("Third Triangle Base:" , t3.getBase())
print("Third Triangle Height:" , t3.getHeight())
print("Third Triangle area:" ,t3.area()) 

```

```python
# Task 7

class Dolls:
  def __init__(self, name, price, flag=False):
    self.name = name
    self.price = price
    self.flag = flag

  def detail(self):
    if self.flag == False:
      return (f'Doll: {self.name}\nTotal Price: {self.price} taka')
    elif self.flag == True:
      return (f'Dolls: {self.name}\nTotal Price: {self.price} taka')

  def __gt__(self, other):
    if self.price > other.price:
      return True
    else:
      return False

  def __add__(self, other):
    price = self.price + other.price
    name = self.name +" "+other.name
    flag = True
    return Dolls(name, price, flag)# call dicchi.ekhane actual val ase ja change
                                   #hobena just constructor e receive hobe.
obj_1 = Dolls("Tweety", 2500)
print(obj_1.detail())
if obj_1 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")

print("=========================")
obj_2 = Dolls("Daffy Duck", 1800) 
print(obj_2.detail())
if obj_2 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")

print("=========================")
obj_3 = Dolls("Bugs Bunny", 3000)
print(obj_3.detail())
if obj_3 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")

print("=========================")
obj_4 = Dolls("Porky Pig", 1500)
print(obj_4.detail())
if obj_4 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")

print("=========================")
obj_5 = obj_2 + obj_3
print(obj_5.detail())
if obj_5 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
```

```python
# Task 8

class Coordinates:                           # num1,num2
  def __init__(self, num1, num2):            # 1,2
    self.num1 = num1                         # 3,4 
    self.num2 = num2
    

  def __sub__(self, otherSub):
    self.sub1 = self.num1 - otherSub.num1   # 1-3 ...  =-2
    self.sub2 = self.num2 - otherSub.num2   # 2-4 ...  =-2
    return Coordinates(self.sub1, self.sub2)

  def __mul__(self, otherMul):
    self.mul1 = self.num1 * otherMul.num1   # 1*3 ...  =3
    self.mul2 = self.num2 * otherMul.num2   # 2*4 ...  =8
    return Coordinates(self.mul1, self.mul2)


  def __eq__(self, otherEq): #mul aar sub tai check korte hobe but ami mul sub ke return kore num ei pathey disi tai ekhaneo normal operator overloading e ja kori oirokom korlei hobe mane self.num er shathe other.num check korbo.
    if self.num1 == otherEq.num1 and self.num2 == otherEq.num2:
      return (f'The calculated coordinates are the same.')
    else:
      return ('The calculated coordinates are NOT the same.')


  def detail(self):
    return self.num1,self.num2  # Automatic touple hishebei output ashbe. 
    
         
p1 = Coordinates(int(input()),int(input()))
p2 = Coordinates(int(input()),int(input()))

p4 = p1 - p2 
print(p4.detail())

p5 = p1 * p2
print(p5.detail())

point_check = (p4 == p5)
print(point_check)
```

```python
# Quiz 3(previous)

# Design a class named Coordinates with the required constructor to produce the following output.
# Use suitable access modifiers with the variables so that they can only be accessed from that class.
# Write getX(), getY() and getZ() methods to access the variables.
# Handle the operator overloading by using corresponding special methods.


class Coordinates:
    def __init__(self, num1, num2, num3):
        self.__num1 = num1
        self.__num2 = num2
        self.__num3 = num3

    def __mul__(self, other):
        self.__other = other

        self.__resMmul1 = self.__num1 * other.__num1
        self.__resMmul2 = self.__num2 * other.__num2
        self.__resMmul3 = self.__num3 * other.__num3
        return Coordinates(self.__resMmul1, self.__resMmul2, self.__resMmul3)

    def __truediv__(self, other):
        self.__other = other

        self.__resDiv1 = self.__num1 / other.__num1
        self.__resDiv2 = self.__num2 / other.__num2
        self.__resDiv3 = self.__num3 / other.__num3
        return Coordinates(self.__resDiv1, self.__resDiv2, self.__resDiv3)

    def getX(self):
        return self.__num1

    def getY(self):
        return self.__num2

    def getZ(self):
        return self.__num3


c1 = Coordinates(3, 4, 5)
c2 = Coordinates(1, 3, 4)
calc1 = c1 * c2
print(calc1.getX())
print(calc1.getY())
print(calc1.getZ())
calc2 = c1 / c2
print(calc2.getX())
print(calc2.getY())
print(calc2.getZ())
```
