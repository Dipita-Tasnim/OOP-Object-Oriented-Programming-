```python
# Task 2 (output perfectly ashenai)-----[solved]

class RealNumber:
  def __init__(self, number=0):
    self.number = number
  def __add__(self, anotherRealNumber):
    return self.number + anotherRealNumber.number
  def __sub__(self, anotherRealNumber):
    return self.number - anotherRealNumber.number
  def __str__(self):
    return str(self.number)
    

class ComplexNumber(RealNumber):
  def __init__(self, real, im):
    if type(real) == int:
      super().__init__(real) # parent er shathe milay real er jagay number dile-
    else:                    #    |      |         |        |
      super().__init__(real.number) #ekhane obj pass koray (number.number) hoto.
    self.im = im

  def __add__(self, other):
    real_number = self.number + other.number
    imaginary_number = self.im + other.im

    return ComplexNumber(real_number, imaginary_number)

  def __sub__(self, other):
    real_number = self.number - other.number
    imaginary_number = self.im - other.im

    return ComplexNumber(real_number, imaginary_number)   #ekdom shuru te constructor
                                                        # er self.number aar self.im e
  def __str__(self):                                  # pathay dilam.
    if self.im < 0:
      return f'{str(self.number)} - {int(str((-1)*(self.im)))}i'
    else:         # Not self.real!
      return f'{str(self.number)} + {str(self.im)}i'

r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
# Task 5 

class Exam:
  def __init__(self,marks):
    self.marks = marks
    self.time = 60
        
  def examSyllabus(self):
    return "Maths , English"
  def examParts(self):
    return "Part 1 - Maths\nPart 2 - English\n"
    


class ScienceExam(Exam):
  def __init__(self, marks, time, *args):
    super().__init__(marks)
    self.time = time
    self.lst = []
    self.count = 2

    for i in args:
      self.lst.append(i)
      self.count += 1
    
  def __str__(self):
    return f'Marks: {str(self.marks)} Time: {str(self.time)} minutes Number of parts: {str(self.count)}'    

  def examSyllabus(self):
    var1 = super().examSyllabus()
    for i in self.lst:
      var1 = var1 +' , ' + i # Normally + diye parent aar nijer jinish jora deya hoy.
    return var1  # Ekhane slicing lagtese na karon comma ta i er aagei ashtese.
                # kono kichur shathe comma diye i concat korte thakle slicing kore
                # last er comma baad deyar jhamela thake na.
  def examParts(self):
    var = super().examParts()
    partCount = 3

    for i in self.lst:
      var = var + f'Part {str(partCount)} - {i}\n' # Normally + diye parent aar nijer jinish jora deya hoy.
      partCount = partCount + 1
    return var



engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================') 
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())

```

```python
# Task 6

class Shape3D:

  pi = 3.14159
  def __init__(self, name = 'Default', radius = 0):
    self._area = 0
    self._name = name
    self._height = 'No need'
    self._radius = radius   

  def calc_surface_area(self):
    return 2 * Shape3D.pi * self._radius

  def __str__(self):
      return "Radius: "+str(self._radius)




class Sphere(Shape3D):
  def __init__(self, name, radius, height = 'No need'):
    super().__init__(name, radius)
    self._height = height
    print(f'Shape name: {self._name}, Area Formula:  4 * pi * r * r')
   
  def calc_surface_area(self):
    # self._area = 4 * Shape3D.pi * self._radius * self._radius #without inheritance
    self._area = super().calc_surface_area() * 2 * self._radius #with inheritance

  def __str__(self):
    return f'Radius: {str(self._radius)}, Height: {str(self._height)}\nArea: {str(self._area)}'




class Cylinder(Shape3D):
  def __init__(self, name, radius, height):
    super().__init__(name, radius)
    self._height = height
    print(f'Shape name: {self._name}, Area Formula:  2 * pi * r * (r+h)')

  def calc_surface_area(self):
    self._area = 2 * Shape3D.pi * self._radius * (self._radius + self._height)

  def __str__(self):  
    return f'Radius: {str(self._radius)}, Height: {str(self._height)}\nArea: {str(self._area)}'



sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
# Quiz 4

class Booking:
    item_list = []

    def __init__(self, title, id):
        self.title = title
        self.id = id
        self.price = 0

    def get_title_id(self):
        return self.title + " | " + self.id 
    
class RoomBooking(Booking): 
    def __init__(self, title, id, type, price):
        super().__init__(title, id)
        self.type = type
        self.price = price
        super().item_list.append(id)
    
    def set_type(self, val):
        self.type = val

    def set_price(self, val):
        self.price = val

    def __str__(self):
        return super().get_title_id() + " | " + self.type + " | " + self.price

    @classmethod
    def from_string_with_hash(cls, string_with_hash):
        title, id, type, price = string_with_hash.split('#')
        return cls(title, id, type, price)

    @classmethod
    def is_available(cls, id):
        if(id in cls.item_list):
            return 'False' # string na dileo hoy.
        else:
            return 'True'

print(len(RoomBooking.item_list))
print("##########")
rb1 = RoomBooking("Room", "ROOM1001", "Single", "8000")
print(rb1)
print("##########")
print(len(RoomBooking.item_list))
print("##########")
print(RoomBooking.is_available("ROOM1001"))
print("##########")
print(RoomBooking.is_available("ROOM1002"))
print("##########")
rb2 = RoomBooking.from_string_with_hash("Room#ROOM1002#Double#15000")
print(rb2)
print("##########")
print(RoomBooking.is_available("ROOM1002"))
print("##########")
print(len(RoomBooking.item_list))
print("##########")
rb1.set_type("Double")
rb1.set_price("15000")
print(rb1)
```
