```python
# Task 5

class Employee:
  def __init__(self, name, workingPeriod):
    self.name = name
    self.workingPeriod = workingPeriod

  @classmethod
  def employeeByJoiningYear(cls, name, year):
    workTime = 2022 - year
    return Employee(name, workTime)

  @staticmethod
  def experienceCheck(workingPeriod, gender):
    if workingPeriod < 3 and gender == 'male':
      return ('He is not experienced')

    elif workingPeriod < 3 and gender == 'female':
      return ('She is not experienced')

    elif workingPeriod >= 3 and gender == 'male':
      return ('He is experienced')

    elif workingPeriod >= 3 and gender == 'female':
      return ('She is experienced')


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
# Task 8

import math
class Cylinder:
    radius = 5
    height = 18
    def __init__(self,up_radius, up_height):
        self.up_radius = up_radius
        self.up_height = up_height

        print(f'Default radius={Cylinder.radius} and height={Cylinder.height}')
        print(f'Updated:radius={self.up_radius} and height={self.up_height}')

        Cylinder.radius = self.up_radius#Ei part tai output dekhe bujhar jinish
        Cylinder.height = self.up_height

    @staticmethod
    def area(up_radius, up_height):
        area = (2*math.pi*float(up_radius)**2) + (2*math.pi*float(up_radius)*float(up_height))
        print('Area:', area)

    @staticmethod
    def volume(up_radius, up_height):
        volume = math.pi*float(up_radius)**2*float(up_height)
        print('Volume:', volume)

    @classmethod
    def swap(cls, height, radius):
        return Cylinder(radius, height) #Cylinder na likhe cls likhleo hoy.

    @classmethod
    def changeFormat(cls, info):
        radius, height = info.split('-')
        return Cylinder(float(radius), float(height))


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 10

class SultansDine:

  branch_num = 0
  total_sell = 0
  info = []

  def __init__(self, b_name):
    self.branch = b_name
    SultansDine.branch_num += 1
    self.sell = 0
    self.percentage = 0
    SultansDine.info.append(self)

  @classmethod
  def details(cls):
    print(f'''Total Number of branch(s): {cls.branch_num}
Total Sell: {cls.total_sell} Taka''')
    if cls.branch_num != 0:
      for iter in cls.info:
        print(f'''Branch Name: {iter.branch}, Branch Sell: {iter.sell} Taka
Branch consists of total sell's: {"{0:5.2f}".format(iter.sell / cls.total_sell * 100)}% ''')

  def sellQuantity(self, quant):
    if quant < 10:
      self.sell = quant * 300
    elif quant < 20:
      self.sell = quant * 350
    else:
      self.sell = quant * 400
    SultansDine.total_sell += self.sell

  def branchInformation(self):
    print(f'''Branch Name: {self.branch}
Branch Sell: {self.sell} Taka''')


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
